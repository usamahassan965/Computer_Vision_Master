### **Loading Images**

```
# Load an image using 'imread' specifying the path to image
# 1 is default flag - for colored images reads in BGR 8-bit format.
image = cv2.imread('image.jpg',1)
print(image.shape)
```

```
img_gray = cv2.imread("image.jpg",0) # 0 for grayscale
print(img_gray.shape)
```

```
img_unchanged = cv2.imread("image.jpg",-1) # -1 loads the image as is (including the alpha channel if present)
print(img_unchanged.shape)
```

### **Displaying Images**

- Opencv loads image in the format of BGR while the matplotlib library processes it in RGB format.

**Note:**
RGB pixel format is quit standard among MATLAB or other image processing liberaies, but OpenCV uses BGR pixel format.

```
#we will be using this function to display image later in notebook
def displayImage(image,title):
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```
displayImage(image,'RGB')
```

```
#showing an image
displayImage(img_gray,"Gray")
```

```
#showing an image - BGR default format
plt.imshow(img_unchanged)
plt.title('Unchanged')
plt.show()
```

### **Saving Images**

```
# Simply use 'imwrite' specificing the file name and the image to be saved
cv2.imwrite('output.jpg', image)
```

```
# Or save it as a PNG (Portable Network Graphics) which is a Lossless bitmap image format
# It takes greater space than that of saving in jpg format
cv2.imwrite('output.png', image)
```
